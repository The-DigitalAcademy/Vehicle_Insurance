```python
gender = insurance['gender'].unique()

gender

gender = pd.DataFrame(gender, columns=['gender'])
gender

gender = gender.reset_index()
gender

gender.rename(columns={'index':'gender_id'}, inplace=True)
gender

# PostgreSQL Connection
DATABASE_URL = "postgresql://dylan:super123duper@129.232.211.166:5432/vehicle_insurancedb"

# Create a database connection
engine = create_engine(DATABASE_URL)

conn = engine.connect()

conn.begin()

# Connect to PostgreSQL
with engine.connect() as conn:
    for _, row in gender.iterrows():
        sql = text("""
            INSERT INTO gender (gender_id, gender)
            VALUES (:gender_id, :gender)
            ON CONFLICT (gender_id) DO NOTHING
        """)
        conn.execute(sql, {"gender_id": row["gender_id"], "gender": row["gender"]})
    conn.commit() 
# Commit and Close Connection
conn.close()

gender



```

```python
def putidforage(x):
    if x == "> 2 Years":
        age = 0
    elif x == '1-2 Year':
        age = 1
    elif x == "< 1 Year":
        age = 2
    else:
        age = "old"
    
    return age
    
insurance['va_id'] = insurance['vehicle_age'].apply(putidforage)    

```

```python
def putidfordamage(x):
    if x == "Yes":
        damage = 0
    elif x == 'No':
        damage = 1
    else:
        damage = "unknown"
    
    return damage
    
insurance['vd_id'] = insurance['vehicle_damage'].apply(putidfordamage)

```

```python
def putidforgender(x):
    if x == "Male":
        gender = 0
    elif x == 'Female':
        gender = 1
    else:
        gender = "unknown"
    
    return gender
    
insurance['gender_id'] = insurance['gender'].apply(putidforgender)

```

```python
# Create a dictionary for region_code -> rc_id mapping
region_code_dict = dict(zip(region_code['region_code'], region_code['rc_id']))

# Function to map region_code to rc_id
def putidforregion(x):
    return region_code_dict.get(x, "unknown")  

# Apply function to assign rc_id
insurance['rc_id'] = insurance['region_code'].apply(putidforregion)

insurance
```

```python
 # Create a dictionary for policy_channel -> pc_id mapping
policy_channel_dict = dict(zip(policy_channel['policy_channel'], policy_channel['pc_id']))

# Function to map region_code to pc_id
def putidforpolicy(x):
    return policy_channel_dict.get(x, "unknown")  

# Apply function to assign pc_id
insurance['pc_id'] = insurance['policy_channel'].apply(putidforpolicy)

insurance
```
